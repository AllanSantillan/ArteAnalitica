```python
#plot
import matplotlib.pyplot as plt
colA=dt['Small Bags']
colB=dt['Large Bags']
colC=dt['XLarge Bags']
colD=dt["Total Bags"]
#colD=dt["user_created"]
myD=[colA,colB,colC,colD]
f=plt.figure(figsize=(6,4))
ax=f.add_axes([0,0,1,1])
bp=ax.boxplot(myD)
plt.show()
```

```python
n_bins=20
colors=["green", "blue", "red"]
plt.hist(myD,n_bins,density = True,
         histtype="bar",
         color = colors,
         label = colors)
plt.legend(prop={"size":12})
plt.title("Histograma de archivo avocado\n\n",
         fontweight="bold")
plt.show()
```
